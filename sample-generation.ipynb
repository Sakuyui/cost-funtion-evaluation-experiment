```python
import random
import numpy as np
from functools import reduce, cache
```

```python

global solution_dictionary 
solution_dictionary = dict()

def record_and_evenly_return(key_identification, data):
    global solution_dictionary
    solution_dictionary[key_identification] = data
    return data

def generate_one_sample_k_cores(neuron_count, k):
        global solution_dictionary
        key_identification = "%d#%d" % (neuron_count, k)
        
        if key_identification in solution_dictionary:
            return solution_dictionary[key_identification]
        if k <= 0:
            return record_and_evenly_return([])
        if k == 1:
            return record_and_evenly_return([neuron_count])
        if k == 2:
            l = []
            for i in range(1, neuron_count):
                sub_solution_k_minus_1 = generate_one_sample_k_cores(i, k - 1)
                t = [[neuron_count - i] + [l] for l in sub_solution_k_minus_1]
                l.append(t)
            return record_and_evenly_return(list(reduce(lambda x,y: x + y, l)))
        if k > 2:
            l = []
            for i in range( k - 1, neuron_count):
                sub_solution_k_minus_1 = generate_one_sample_k_cores(i, k - 1)
                t = [[neuron_count - i] + l for l in sub_solution_k_minus_1]
                l.append(t)
            return record_and_evenly_return(list(reduce(lambda x,y: x + y, l)))

def generate_samples(neuron_counts_each_application_vertex):
    import random
    application_vertex_count = len(neuron_counts_each_application_vertex)
    total_neuron_count = sum(neuron_counts_each_application_vertex)
    
    samples = []
    for k in range(1, total_neuron_count + 1):
        samples.append(generate_one_sample_k_cores(total_neuron_count, k))
        if k == 3:
            break
    return list(reduce(lambda x,y: x + y, samples))
```
